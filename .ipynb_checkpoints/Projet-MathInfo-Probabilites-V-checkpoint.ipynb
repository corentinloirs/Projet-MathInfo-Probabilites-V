## Questions théoriques

###### Question 1

On cherche à simuler un vecteur gaussien Z, qui est donc de carré intégrable. On voudrait simuler l'espérance conditionelle du vecteur gaussien en calculant la moyenne empirique de simulations conditionelles. Pour justifier ce choix, on va utiliser la loi des grands nombres.

On va noter $Z_i$ la variable aléatoire associée à la $i^{eme}$ simulation du vecteur $Z$. Toutes ces variables sont bien des vecteur gaussiens de même espérance $\mu$ et matrice de covariance $\Sigma$ que $Z$, et ils sont de carré intégrable (propriété des vecteurs gaussiens).

On peut donc appliquer la loi des grands nombres, qui stipule que:

En notant $M_n$ le vecteur $\frac{X_1+...+X_n}{n}$, on a $M_n$ converge en moyenne et presque sûrement vers $\mu$. On a donc simulé l'espérance par la moyenne des résultats empiriques. Il suffit ensuite de remplacer $\mathbb{E}$ par l'espérance conditionelle recherchée, d'ou le résultat.

###### Question 2

Pour cette question, on se réfère directement aux pages 12 et 13 du cours Probabilité IV.
On note $B = (z(x_{j_1}),z(x_{j_2}),...,z(x_{j_n}))$ le vecteur correspondant aux coordonées fixées observées, et $Y$ le vecteur des N-n autres coordonées qu'on cherche à estimer.

On exprime donc d'après Probabilités IV

$f_{Y\,|B = b}(z)\ =\ \frac{1}{(2\pi)^{(N-n)\,/\,2}\sqrt{det(CS_Y)}}exp\,(\frac{-1}{2}(y-\psi(b))^tCS_Y^{-1}(y-\psi(b)))$

###### Question 3

D'après le paragraphe en haut de la page 12 du chapitre Probabilité V (qui renvoie à l'exercice Combinaisons linéaires de variables aléatoires Gaussiennes indépendantes du cours Probabilités II) : si $Y$ suit une loi normale centrée réduite, alors $R Y\,+\,m$ suit une loi normale d’espérance $m$ et de matrice de covariance $R$.

###### Question 4

On sait simuler une variable suivant une loi normale centrée réduite d'après la méthode de *Box-Muller* décrite dans le chapitre Probabilités V.
Dès lors, on va simuler $n^2$ variables aléatoires de la sorte. On obtient une matrice de taille $n^2$ qu'on va multiplier par la matrice de covariance C associée au problème ($R=C$) et on utilise la transformation de la question 2 pour obtenir l'espérance conditionelle.

On construit ainsi un vecteur $X_i$. On répète la simulation $k$ fois (avec k grand), et on obtient d'après la question 1 en moyenne une bonne estimation de la longueur de cable.

## Partie informatique

Importations :

```python
#Chargement de dépendances
import numpy as np 
import matplotlib.pyplot as plt
from math import *
from random import uniform
#Discrétisation 
A=0 
B=500 
N=101 
#Nombre de points de discrétisation 
Delta = (B-A)/(N-1) 
discretization_indexes = np.arange(N) 
discretization = discretization_indexes*Delta #Paramètres du modèle
mu=-5 
a = 50 
sigma2 = 12
#Données
observation_indexes = [0,20,40,60,80,100] 
depth = np.array([0,-4,-12.8,-1,-6.5,0])

unknown_indexes=list(set(discretization_indexes)-set(observation_indexes))
```

###### Question 1

Le *try* dans cette fonction permet de différencier le cas matriciel du cas scalaire

```python
def cov(d,a,b):
    try : 
        n=len(d[:,1])
        C=np.zeros((n,n))
        for i in range (0,n):
            for j in range (0,i+1):
                C[i,j]=b*exp(-abs(d[i,j])/a)
                C[j,i]=C[i,j]
        return C
    except:
        return b*exp(-abs(d)/a)
```

```python
#la coordonnée i,j : distance entre xi et xj 
DIST = np.zeros((N,N))
for i in range (0,N):
    for j in range (0,i+1):
        DIST[i,j]=abs(discretization[i]-discretization[j])
        DIST[j,i]=DIST[i,j]

COV=cov(DIST,a,sigma2)
```

##### Question 4

```python
#entre les observations : 
observation=[i*Delta for i in observation_indexes]
n1=len(observation)
DISTobs=np.zeros((n1,n1))
for i in range (0,n1):
    for j in range (0,i+1):
        DISTobs[i,j]=abs(observation[i]-observation[j])
        DISTobs[j,i]=DISTobs[i,j]

COVobs=cov(DISTobs,a,sigma2)
print(COVobs)

#entre les inconnues : 
unknown=[i*Delta for i in unknown_indexes]
n2=len(unknown)
DISTunk=np.zeros((n2,n2))
for i in range (0,n2):
    for j in range (0,i+1):
        DISTunk[i,j]=abs(unknown[i]-unknown[j])
        DISTunk[j,i]=DISTunk[i,j]

COVunk=cov(DISTunk,a,sigma2)
print(COVunk)

#entre les observations et les inconnues :
COVobun=np.zeros((n2,n1))
for i in range(0,n2):
    for j in range(0,n1):
        d=abs(unknown[i]-observation[j])
        COVobun[i,j]=cov(d,a,sigma2)

        
print(COVobun)
```

##### Question 5

On ne représnete pas l'espérance ici car on la plotera à la question 7.

```python
Mobs=[mu]*n1
Munk=[mu]*n2
COVobsinv = np.linalg.inv(COVobs)
ESP=Munk+np.dot(COVobun,np.dot(COVobsinv,(depth-Mobs)))
```

##### Question 6

```python

COVcond=COVunk-np.dot(COVobun,np.dot(COVobsinv,COVobun.T))
diag=np.diagonal(COVcond)

plt.figure() 
plt.plot(unknown,diag)
plt.grid()
plt.title("Diagonale de la matrice de covariance, en fonction de l'abscisse")
plt.xlabel("Abscisse")
plt.ylabel("Ordonnée")
plt.show()
```

##### Question 7

```python
#simulation d'un vecteur gaussien aléatoire Y aux composantes centrées réduites aléatoires
L=np.linalg.cholesky(COVcond)
Y=[]
for i in range (len(ESP)):
    u=uniform(0,1)
    v=uniform(0,1)
    Y+=[((-2*np.log(u))**.5)*cos(2*pi*v)]
Z=ESP+np.dot(L,Y)

plt.figure()
plt.scatter(unknown,Z,c="blue",label="Profondeurs simulées")
plt.scatter(unknown,ESP,c="green",label="Espérance de la profondeur")
plt.scatter(observation,depth,c="red",label="Profondeurs mesurées")
plt.legend()
plt.xlabel("Abscisses")
plt.ylabel("Profondeur")
plt.title("Profondeur obtenue par simulation et profondeur donnée par l'espérance")
plt.grid()
plt.show()
```

On obtient bien une distribution autour de l'espérance de la profondeur

##### Question 8

```python
def longueur(profondeur, pas):
    l=0
    for i in range(len(profondeur)-1):
        l+=sqrt(pas**2+(profondeur[i]-profondeur[i+1])**2)
    return l
```

##### Question 9

```python
def simulation(ESPER):
    Y=[]
    for i in range (len(ESPER)):
        u=uniform(0,1)
        v=uniform(0,1)
        Y+=[((-2*np.log(u))**.5)*cos(2*pi*v)]
    Z=ESPER+np.dot(L,Y)
    abscisse=unknown.copy()
    z=Z.tolist()
    esp=ESPER.tolist()
    for i in range(len(observation)):
        j=0
        while j<len(abscisse) and abscisse[j]<observation[i]:
            j+=1
        abscisse.insert(j,observation[i])
        z.insert(j,depth[i])
        esp.insert(j,depth[i])
    Z=np.asarray(z)
    ESPER=np.asarray(esp)
    return Z,ESPER
```

Remarque: on renvoie Z et l'espérance par pur soucis de simplification des algorithmes suivants

```python
longueuresp=longueur(simulation(ESP)[1],Delta)
LONGUEUR=[]
for i in range(100):
    Z=simulation(ESP)[0]
    LONGUEUR+=[longueur(Z,Delta)]
longu=sum(LONGUEUR)/len(LONGUEUR)
print("différence entre la longueur moyenne obtenue après 100 simulations et la longueur de l'espérance:",longu-longueuresp," m")
```

On obtient une différence tendant vers 20, qu'on n'a a priori du mal à s'expliquer: d'après la loi des grands nombres, on devrait avoir une différence qui tend vers 0 quand le nombre de simulations tend vers l'infini. En fait, ici, on reste à n "faible" (n = 100). On peut donc justifier une différence significative de 20 mètres.

##### Question 10

On plot directement la différence plutôt que les $M_n$ (car c'est avant tout ça qui nous intéresse). L'histogramme de la question 11 donnera un aperçu de la répartition des $M_n$.

```python
LONGUEURn=[]
LONGUEUR=[]
for i in range(1,100):
    Z=simulation(ESP)[0]
    LONGUEUR+=[longueur(Z,Delta)]
    LONGUEURn+=[sum(LONGUEUR)/len(LONGUEUR)]
ERRn=[abs(x-longueuresp) for x in LONGUEURn]
n=np.arange(1,100,1)
plt.figure()
plt.plot(n,ERRn)
plt.grid()
plt.xlabel("n,nombre de simulations effectuées")
plt.ylabel("différence")
plt.title("Différence entre la longueur de l'espérance et la longueur moyenne en fonction du nombre de simulations" )
plt.show()
```

On remarque que cet histogramme ne ressemble pas vraiment à une gaussienne

##### Question 12

```python
#Une première méthode consiste à simplement garder le "milieu" de l'échantillon obtenu sur l'ensemble des simulations.
#On fait suffisement confiance à l'échantillon pour considérer qu'il est représentatif

LONGUEURBORNE = []
for i in range(100):
    Z = simulation(ESP)[0]
    LONGUEURBORNE.append(longueur(Z,Delta))
LONGUEURBORNE.sort()
borne = len(LONGUEURBORNE)
borneinf = 0.025*borne
bornesup = 0.975*borne
print(f"L'intervalle de confiance à 95% par cette méthode est [{LONGUEURBORNE[int(borneinf)]}, {(LONGUEURBORNE[int(bornesup)])}]")
```

```python
#Une deuxième méthode plus précise consiste à utiliser la formule connue depuis le lycée, obtenue par application de l'inégalité de Bienaymé-Tchebychev
#https://fr.wikipedia.org/wiki/Intervalle_de_confiance, cf. "Formulation de l'intervalle de confiance autour d'une moyenne observée x
#avec un écart type observé s sur un échantillon de taille n."

LONGUEURBORNE = [] #pas nécessaire mais comme ça on s'évite tout problème de réécriture. On gardera ce LONGUEURBORNE pour la suite.
for i in range(100):
    Z = simulation(ESP)[0]
    LONGUEURBORNE.append(longueur(Z,Delta))
moyennelongueur = np.mean(LONGUEURBORNE)
eclongueur = sqrt(np.var(LONGUEURBORNE))
borneinf = moyennelongueur - 1.96*(eclongueur/10)
bornesup = moyennelongueur + 1.96*(eclongueur/10)
print(f"L'intervalle de confiance à 95% par cette méthode est [{borneinf}, {bornesup}]")
```

On reste extrêmement proche de l'espérance avec cet "offset" de 20 mètres pour les deux méthodes. On remarque que la seconde méthode, nettement moins grossière, donne un résultat plus fin (la plage de valeur est moins grande).

##### Question 13

```python
#On mesure simplement la proportion de longueurs supérieures à 525 mètres en moyenne sur 100 simulations
acces =  0
for elt in LONGUEURBORNE:
    if elt >= 525:
        acces +=1
print(f"La probabilité que la longueur du cable soit supérieure à 525 mètres est d'environ {acces/len(LONGUEURBORNE)}")
```
